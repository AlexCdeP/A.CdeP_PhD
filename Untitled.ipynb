```python
#Open file from 07/08/2018

data_f = df.current()
fname = data_f.filename
```

```python
#Access spectra
filelist = data_f['OceanOpticsSpectrometer']
```

```python
#Extract wavelengths and DF and BF background from first spectrum in HDF5 file. DF and BF backgrounds have
#different integration times

wavelength = filelist[filelist.keys()[0]].attrs["wavelengths"]
background_DF = filelist[filelist.keys()[0]].attrs["background"]


#For now simply estimating the background for BF since forgot to do it for 100ms int time!
background_BF = background_DF/10 #Since integration time here was 1000ms
```

```python
for i in range(0, len(filelist.keys())):
    if filelist.keys()[i] == 'Mirror_BF_100msInt_0':
        Mir_val = i
    elif filelist.keys()[i] == 'Au_onlyBF_0':
        Au_val_1 = i
        Au_val_2 = i + 1
        Au_val_3 = i + 2
        Au_val_4 = i + 3
        Au_val_5 = i+4
        Au_val_6 = i+5
    else:
        pass

```

```python
#Extract mirror spectrum from dataset
mir_R = filelist[filelist.keys()[Mir_val]].value
Au_R_1 = filelist[filelist.keys()[Au_val_1]].value
Au_R_2 = filelist[filelist.keys()[Au_val_2]].value
Au_R_3 = filelist[filelist.keys()[Au_val_3]].value
Au_R_4 = filelist[filelist.keys()[Au_val_4]].value
Au_R_5 = filelist[filelist.keys()[Au_val_5]].value
Au_R_6 = filelist[filelist.keys()[Au_val_6]].value

Au_R_av = (Au_R_1 +Au_R_2+Au_R_3+Au_R_4+Au_R_5+Au_R_6)/6

mir_R_corr = mir_R - background_BF
Au_R_corr = Au_R - background_BF
Au_R_av_corr = Au_R_av - background_BF
```

```python
#Looking better!
plt.plot(wavelength, mir_R_corr)
plt.plot(wavelength, Au_R_corr)
#plt.plot(wavelength, Au_R_av_corr)
```

```python
Au_absorption_corr = Au_R_corr/mir_R_corr
```

```python
fig1 = plt.figure(1)
ax1 = fig1.add_subplot(111)
#ax1.plot(wavelength, Au_absorption)
ax1.set_xlim(400, 1000)
ax1.set_ylim(0,1.1)

Au_R_smooth = sig.savgol_filter(Au_absorption_corr,23,3)

ax1.plot(wavelength, Au_absorption_corr)

np.savetxt("C:/Users/alexc/OneDrive - University Of Cambridge/Documents/PhD Project/Projects/Fabry-Perot hBN/Experimental_Au_R_smoothed.txt", Au_R_smooth, delimiter = "\t")
np.savetxt("C:/Users/alexc/OneDrive - University Of Cambridge/Documents/PhD Project/Projects/Fabry-Perot hBN/All_wls.txt", wavelength, delimiter = "\t")
```

```python
Au_absorption_av = Au_R_av_corr/mir_R_corr
fig1 = plt.figure(1)
ax1 = fig1.add_subplot(111)
#ax1.plot(wavelength, Au_absorption)
ax1.set_xlim(400, 1000)
ax1.set_ylim(0,1.1)

#Au_R_smooth_av = sig.savgol_filter(Au_absorption_av,23,3)

np.savetxt("C:/Users/alexc/OneDrive - University Of Cambridge/Documents/PhD Project/Projects/Fabry-Perot hBN/Experimental_Au_R_av.txt", Au_absorption_av, delimiter = "\t")
np.savetxt("C:/Users/alexc/OneDrive - University Of Cambridge/Documents/PhD Project/Projects/Fabry-Perot hBN/Experimental_Au_R_av_smoothed.txt", Au_R_smooth_av, delimiter = "\t")

ax1.plot(wavelength, Au_absorption_av)
```
