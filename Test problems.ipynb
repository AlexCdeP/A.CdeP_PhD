```python
# Trying another approach to solving the rucksack problem
weights = [3, 6, 1, 9, 13, 14]
values = [4, 9, 1, 3, 20, 13]
max_wt = 20
n = len(weights) 
weights.insert(0,0) # improves alignment of arrays with K
values.insert(0,0)


K = [0]*(max_wt+1) # This will contain all of the maximum weights for n items with a capacity of w, 
#which is the index of K

K[0] = 0 # This is because the first weight capacity of 0 can contain 0 items with 0 max value

# Iterate through all weight capacities increasing by 1 weight unit each time and use the result later on 
for w in range(1, max_wt+1):
    #set the current value for n items with capcity w to 0
    max_cap_val = 0
    # Now need to iterate through all the weight values to check if there is a value less than the current w
    i=0
    for i in range(1, n):
        wi = weights[i]
        vi = values[i]
        # now check if the current w is greater than or equal to a weight available
        if w >= wi:
            # now if we take this item there will be a weight of w-wi remaining so we need to take the max value for
            # w-wi and add it to the value for the current item
            new_cap_val = K[w-wi] + vi
            # Now need to check if this is greater than the previous combination (if there was one)
            # If so this is the new maximum capacity for weight w
            if new_cap_val > max_cap_val:
                max_cap_val = new_cap_val
    # Last thing to do is to populate K so the maximum value for weight w is contained here
    K[w] = max_cap_val

#That should be it
max_value = K[-1]        
```
